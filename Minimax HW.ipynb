```python
def minimax(skill1, skill2):
    score = 0
    turn = 0
    playerturn = 0
    score = [0,0,0,0,0,0,0,0,0]
    win = 0
    
  
    while(turn < 10):
        #first turn
        if(turn == 0):
            print("\n")
            print("player1 turn\n")
            if(skill1 == 1):
                rnd_num_start = random.randint(0, 3)
                if(rnd_num_start == 0):
                    board[0][0] = 'X'
                if(rnd_num_start == 1):
                    board[0][2] = 'X'
                if(rnd_num_start == 2):
                    board[2][0] = 'X'
                if(rnd_num_start == 3):
                    board[2][2] = 'X'
                
            if(skill1 == 0):
                rnd_num_start_check = random.randint(0, 1)
                if(rnd_num_start_check == 0):
                    rnd_num_start_randompos = random.randint(0, 8)
                    if(rnd_num_start_randompos == 0):
                        board[0][0] = 'X'
                    if(rnd_num_start_randompos == 1):
                        board[0][1] = 'X'
                    if(rnd_num_start_randompos == 2):
                        board[0][2] = 'X'
                    if(rnd_num_start_randompos == 3):
                        board[1][0] = 'X'
                    if(rnd_num_start_randompos == 4):
                        board[1][1] = 'X'
                    if(rnd_num_start_randompos == 5):
                        board[1][2] = 'X'
                    if(rnd_num_start_randompos == 6):
                        board[2][0] = 'X'
                    if(rnd_num_start_randompos == 7):
                        board[2][1] = 'X'
                    if(rnd_num_start_randompos == 8):
                        board[2][2] = 'X'
                if(rnd_num_start_check == 1):
                    rnd_num_start = random.randint(0, 3)
                    if(rnd_num_start == 0):
                        board[0][0] = 'X'
                    if(rnd_num_start == 1):
                        board[0][2] = 'X'
                    if(rnd_num_start == 2):
                        board[2][0] = 'X'
                    if(rnd_num_start == 3):
                        board[2][2] = 'X'
        if(turn == 1):
            print("\n")
            print("player2 turn\n")
            if(skill2 == 1):
                if(board[0][0] == 'X'):
                    board[1][1] = 'O'
                if(board[0][2] == 'X'):
                    board[1][1] = 'O'
                if(board[2][0] == 'X'):
                    board[1][1] = 'O'
                if(board[2][2] == 'X'):
                    board[1][1] = 'O'
                if(board[1][1] == 'X'):
                    rnd_num_start = random.randint(0, 3)
                    if(rnd_num_start == 0):
                        board[0][0] = 'O'
                    if(rnd_num_start == 1):
                        board[0][2] = 'O'
                    if(rnd_num_start == 2):
                        board[2][0] = 'O'
                    if(rnd_num_start == 3):
                        board[2][2] = 'O'
                if(board[0][1] == 'X' or board[1][0] == 'X'or board[1][2] == 'X' or board[2][1] == 'X'):
                    rnd_num_start = random.randint(0, 3)
                    if(rnd_num_start == 0):
                        board[0][0] = 'O'
                    if(rnd_num_start == 1):
                        board[0][2] = 'O'
                    if(rnd_num_start == 2):
                        board[2][0] = 'O'
                    if(rnd_num_start == 3):
                        board[2][2] = 'O'
                        
            if(skill2 == 0):
                rnd_num_start_check = random.randint(0, 1)
                if(rnd_num_start_check == 0):
                    rnd_num_start_randompos = random.randint(0, 8)
                    if(rnd_num_start_randompos == 0 and board[0][0] != 'X'):
                        board[0][0] = 'O'
                    if(rnd_num_start_randompos == 0 and board[0][0] == 'X'):
                        board[0][1] = 'O'
                    if(rnd_num_start_randompos == 1 and board[0][1] != 'X'):
                        board[0][1] = 'O'
                    if(rnd_num_start_randompos == 1 and board[0][1] == 'X'):
                        board[0][2] = 'O'
                    if(rnd_num_start_randompos == 2 and board[0][2] != 'X'):
                        board[0][2] = 'O'
                    if(rnd_num_start_randompos == 2 and board[0][2] == 'X'):
                        board[1][0] = 'O'
                    if(rnd_num_start_randompos == 3 and board[1][0] != 'X'):
                        board[1][0] = 'O'
                    if(rnd_num_start_randompos == 3 and board[1][0] == 'X'):
                        board[1][1] = 'O'
                    if(rnd_num_start_randompos == 4 and board[1][1] != 'X'):
                        board[1][1] = 'O'
                    if(rnd_num_start_randompos == 4 and board[1][1] == 'X'):
                        board[1][2] = 'O'
                    if(rnd_num_start_randompos == 5 and board[1][2] != 'X'):
                        board[1][2] = 'O'
                    if(rnd_num_start_randompos == 5 and board[1][2] == 'X'):
                        board[2][0] = 'O'
                    if(rnd_num_start_randompos == 6 and board[2][0] != 'X'):
                        board[2][0] = 'O'
                    if(rnd_num_start_randompos == 6 and board[2][0] == 'X'):
                        board[2][1] = 'O'
                    if(rnd_num_start_randompos == 7 and board[2][1] != 'X'):
                        board[2][1] = 'O'
                    if(rnd_num_start_randompos == 7 and board[2][1] == 'X'):
                        board[2][2] = 'O'
                    if(rnd_num_start_randompos == 8 and board[2][2] != 'X'):
                        board[2][2] = 'O'
                    if(rnd_num_start_randompos == 8 and board[2][2] == 'X'):
                        board[0][0] = 'O'
                        
                        
            
        if(turn > 1):
            #check wins
            if(board[0][0] == 'X'):
                if(board[0][1] == 'X'):
                    if(board[0][2] == 'X'):
                        win = 1
                        
            if(board[0][0] == 'X'):
                if(board[1][0] == 'X'):
                    if(board[2][0] == 'X'):
                        win = 1
                        
            if(board[2][0] == 'X'):
                if(board[2][1] == 'X'):
                    if(board[2][2] == 'X'):
                        win = 1
                        
            if(board[0][2] == 'X'):
                if(board[1][2] == 'X'):
                    if(board[2][2] == 'X'):
                        win = 1
                        
            if(board[0][0] == 'X'):
                if(board[1][1] == 'X'):
                    if(board[2][2] == 'X'):
                        win = 1
                        
            if(board[0][2] == 'X'):
                if(board[1][1] == 'X'):
                    if(board[2][0] == 'X'):
                        win = 1
                        
                        
            if(board[0][0] == 'O'):
                if(board[0][1] == 'O'):
                    if(board[0][2] == 'O'):
                        win = 2
                        
            if(board[0][0] == 'O'):
                if(board[1][0] == 'O'):
                    if(board[2][0] == 'O'):
                        win = 1
                        
            if(board[2][0] == 'O'):
                if(board[2][1] == 'O'):
                    if(board[2][2] == 'O'):
                        win = 2
                        
            if(board[0][2] == 'O'):
                if(board[1][2] == 'O'):
                    if(board[2][2] == 'O'):
                        win = 2
                        
            if(board[0][0] == 'O'):
                if(board[1][1] == 'O'):
                    if(board[2][2] == 'O'):
                        win = 2
                        
            if(board[0][2] == 'O'):
                if(board[1][1] == 'O'):
                    if(board[2][0] == 'O'):
                        win = 2
            
            
            
            
            #player 1 even (X)
            if(playerturn % 2):
                for x in range(0, 2):
                    for y in range(0, 2):
                        if(board[x][y] == 'X'):
                            score[0] = 0
                        if(board[x][y] == 'O'):
                            score[0] = 0
                        if(board[x][y] == ' '):
                            score[0] = 1
                            if(x == 0 and y == 0):
                                if(board[0][2] == 'X' and board[0][1] == 'X')
                                    score[0] = 10
                                if(board[2][0] == 'X')
                                    score[0] = 10
                            
                    
                
                    
            #player 2 odd(O)
            else:
                print("nothing yet")
            
                
        print_board()                    
        turn = turn + 1
        playerturn = playerturn + 1
```
